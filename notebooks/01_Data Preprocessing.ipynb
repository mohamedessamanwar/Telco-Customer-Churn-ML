```python
DATA_PATH = '..\data\WA_Fn-UseC_-Telco-Customer-Churn.csv'
EXPORT_PATH = "..\data\WA_Fn-UseC_-Telco-Customer.pkl"
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import warnings
import pickle
warnings.filterwarnings('ignore')
```

```python
# Read data and print shape
df = pd.read_csv(DATA_PATH)
```

This code converts the 'TotalCharges' column in a pandas DataFrame called 'df' to a numeric data type, with any non-numeric values being replaced with NaN (not a number) using the 'coerce' option. Then it checks for the number of missing values (NaN) in each column of the DataFrame using the 'isnull()' method, and returns the sum of missing values in each column using the 'sum()' method.

```python
df['TotalCharges'] = pd.to_numeric(df.TotalCharges, errors='coerce')
df.isnull().sum()
```

```python
df[df['TotalCharges'].isnull()]
```

To solve the problem of missing values in TotalCharges column, I decided to fill it with the mean of TotalCharges values.

```python
df.fillna(df["TotalCharges"].mean() ,  inplace=True)
```

```python
df[df['TotalCharges'].isnull()]['TotalCharges']
```

```python
df["SeniorCitizen"]= df["SeniorCitizen"].map({0: "No", 1: "Yes"})
df.head()
```
